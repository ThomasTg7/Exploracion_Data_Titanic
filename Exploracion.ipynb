Para comenzar la exploración, es fundamental comprender los datos, para eso primero importaremos las librerías necesarias y luego examinaremos tanto las columnas como los registros a través de una tabla.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("titanic_data/train.csv")
print(df.columns)
print(df.head(10))
```

Primero filtraremos los datos para segmentar a los fallecidos, es decir Survived == 0 y a los supervivientes con Survived == 1.

```python
df_muertos = df[df['Survived'] == 0]
df_sobrevivientes = df[df['Survived'] == 1]
```

A continuación, analizaremos la cantidad de fallecidos por género, utilizando los datos filtrados previamente.

```python
sns.countplot(data=df_muertos, x='Sex')
plt.title('Fallecidos por Sexo')
plt.show()
```

Se puede apreciar que la gran mayoría de las muertes afectaron a los hombres, superando los 400 fallecidos, mientras que en las mujeres la cifra alcanza aproximadamente las 100.

Ahora generaremos un gráfico de la cantidad de muertes por clase y género para facilitar la comparación de los resultados.

```python
sns.countplot(data=df_muertos, x="Sex", hue= "Pclass")
plt.title("Fallecidos por Clase") 
plt.show()
```

Podemos apreciar que tanto en hombres como en mujeres, los fallecidos pertenecen principalmente a la tercera clase, seguida por la segunda y la primera. Si bien en estas últimas dos el número no es tan significativo comparado con la tercera, se podría postular que la probabilidad de sobrevivir era mayor dentro de las dos primeras clases.

Ahora visualizaremos la edad de de los pasajeros segun su clase dentro de la embarcacion

```python
sns.boxplot(data=df, x="Age", hue="Pclass")
```

Podemos observar que la edad de los pasajeros disminuye a medida que bajamos de clase, mientras que en la primera clase muestra una gran variedad de edades, con una tendencia a ser mayores, la tercera clase concentra su volumen principal en pasajeros de los 20 a 30 años.

Ahora visualizaremos las edades de los fallecidos y de los supervivientes

```python
sns.histplot(data=df, x="Age", hue="Survived", element="step")
```

Se puede apreciar que gran parte de los pasajeros se situa entre los 20 y 30 años, tanto en fallecidos como en supervivientes. En cuanto al rango de 0 a 10 años, se observa que hubo una mayor tasa de supervivencia, con aproximadamente 36 supervivientes frente a 20 fallecidos, esta tendencia es aún más marcada entre los 0 y 5 años, esto nos permite postular que los infantes fueron la prioridad al abordar los botes de emergencia.

Respecto a la tercera edad, representan una minoría, observándose que en el rango cercano a los 70 años no hubo supervivientes.

Como conclusión final, podemos afirmar que la mayor tasa de supervivencia corresponde a los infantes y a las mujeres, lo que respalda la idea de que tuvieron prioridad al momento de acceder a los botes de emergencia.

En cuanto a las clases dentro de la embarcacion, podemos observar que la tercera clase fue la más afectada en cuanto a fallecidos, mientras que los pasajeros de primera y segunda clase presentaron una cantidad significativamente menor de fallecidos.

Respecto a la edad, el grupo entre 20 y 30 años fue el más afectado, principalmente porque representaba la mayor parte de los pasajeros, por otro lado, los adultos mayores constituían una minoría y se evidencia que su tasa de supervivencia fue casi nula.
